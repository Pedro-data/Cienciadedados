**Exemplo simples de aplicação do Algoritmo RandomForestClassifier**

O que precisamos saber sobre o RandomForestClassifier
- De forma simples, o algoritmo é capaz de criar estruturas de decisão baseadas em Arvores.
- 
  1_5dq_1hnqkboZTcKFfwbO9A.png

```
#Importação do algoritmo da lib Sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```
#Gerando dados para exemplo 
# X São features que serão entregues para o modelo
# y é o Target
# Ou seja para cada conjunto de 4 Features em X teremos a váriavel reposta que é y 
X, y = make_classification(n_samples=1000, n_features=4,
                          n_informative=2, n_redundant=0,
                          random_state=0, shuffle=False)
```

```
#Visualizando dados de X
print(X)
```

```
print(y)
```

```
 # Mostrando exemplo de X e y
 print(X[:1])
 print(y[:1])
```

```
#Instanciando o algoritmo na váriavel clf
clf = RandomForestClassifier(max_depth=2, random_state=0)
```

```
#Treinando o modelo com método fit()
clf.fit(X, y)
```

```
#Fazendo predição 1
print(clf.predict([[0, 0, 0, 0]]))
```

```
#Fazendo a predição 2
print(clf.predict([[-2, -4, -2, -2]]))
```
