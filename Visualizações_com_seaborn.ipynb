```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```
tips = sns.load_dataset('tips');
```

```
tips.head(4)
```

```
tips.isnull().sum()
#Verificando se há valores nulos no dataset
```

```
sns.relplot(x='size', y='tip', data=tips);
```

```
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips);
```

```
sns.relplot(x='total_bill', y='tip', hue='sex', data=tips);
```

```
sns.pairplot(tips, hue="smoker");
```

```
sns.jointplot("total_bill", "tip", data=tips,
                  kind="reg", truncate=True,
                  xlim=(0, 60), ylim=(0, 12),
                  color="orange", height=6);
```

```
g = sns.jointplot("total_bill", "tip", data=tips, kind="hex", color="orange")
```
