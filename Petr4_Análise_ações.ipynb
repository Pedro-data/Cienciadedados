```
import plotly.graph_objects as go

data = [go.Scatter(
          x=df.Date,
          y=df['Close'])]
py.iplot(data)
```

```
Close = go.Scatter(
    x = df.Date,
    y = df.Close,
    name = 'PETR4 High',
    line = dict(color='#17BECF'),
    opacity = 0.8)

data = [Close]

layout = dict(
    title = 'Análise do ano de 2017',
    title_x = 0.5,
    xaxis = dict(range=['2017-01-01','2017-12-31']),
    yaxis = dict(range=['10','20'])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = 'Manually set range')
```

```
import plotly.graph_objects as go
Close = go.Scatter(
    x = df.Date,
    y = df.Close,
    name = 'PETR4 High',
    line = dict(color='#17BECF'),
    opacity = 0.8)

data = [Close]

layout = dict(title = 'Rangeslider e Rangeselector',
              title_x = 0.5,
              xaxis = dict(
                  rangeselector = dict(
                      buttons = list([
                          dict(count = 1,
                              label = '1m',
                              step='month',
                              stepmode = 'backward'),
                          dict(count = 6,
                              label = '6m',
                              step='month',
                              stepmode = 'backward'),
                          dict(step = 'all')
                      ])
                  ),
                  rangeslider=dict(
                      visible=True
                  ),
                  type = 'date'
                  )
               )

fig = dict(data=data, layout=layout)
py.iplot(fig)
```
